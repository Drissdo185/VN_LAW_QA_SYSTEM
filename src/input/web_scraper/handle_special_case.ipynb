```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Connection': 'keep-alive',
}
```

```python
tem_url = "https://luatvietnam.vn/dan-su/luat-quyen-lap-hoi-1957-van-dang-ap-dung-year-928-d1.html"
reponse = requests.get(tem_url, headers=headers)
soup = BeautifulSoup(reponse.content, 'html.parser')
```

```python
soup
```

```python
section = soup.find_all('div', class_='the-document-body noidungtracuu')
```

```python
content = []
for p in soup.find_all('div', class_='the-document-body noidungtracuu'):
        # Remove empty paragraphs and clean the text
        text = p.get_text(strip=True)
        if text:
            text = re.sub(r'["""]([^"""]*)[""]', r'\1', text)
            text = re.sub(r'\s+', ' ', text.strip())
            # Remove specific HTML entities
            text = text.replace('&nbsp;', ' ')
            # Remove empty parentheses and brackets
            text = re.sub(r'\(\s*\)', '', text)
            text = re.sub(r'\[\s*\]', '', text)
            content.append(text)
    
    # Join all content with newlines
full_content = '\n'.join(content)
```
